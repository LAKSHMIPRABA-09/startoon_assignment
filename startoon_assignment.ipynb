```python
import numpy as np
def find_peaks(signal):
    maxima = []
    minima = []
    for i in range(1, len(signal) - 1):
        if signal[i] > signal[i - 1] and signal[i] > signal[i + 1]:
            maxima.append(i)
        if signal[i] < signal[i - 1] and signal[i] < signal[i + 1]:
            minima.append(i)

    return maxima, minima
```

```python
signal_data_1 = np.loadtxt('Data_1.txt')
signal_data_2 = np.loadtxt('Data_2.txt')
```

```python
maxima_1, minima_1 = find_peaks(signal_data_1)
maxima_2, minima_2 = find_peaks(signal_data_2)
```

```python
print("Maxima for Data_1:", maxima_1)
print("Minima for Data_1:", minima_1)
print("Maxima for Data_2:", maxima_2)
print("Minima for Data_2:", minima_2)
```
